```python
"""
Tạo 1000 dữ liệu kiểu int ngẫu nhiêu gắn với 1000 ngày từ 09/02/2010.
Dùng kiểu pd.Series.

Tính cumsum() và vẽ đồ thị.
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
index = pd.date_range('2010-02-09', periods=1000)
data = np.random.randint(1000, size=1000)
```

```python
result = pd.Series(data, index); result
```

```python
x = np.cumsum(result); x
```

```python
df = pd.DataFrame(result)
df.plot()
```
