```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('dientich.csv', sep=';', header=None, skiprows=4, skipfooter=5, engine='python')[range(6)].set_index(0); data
```

```python
data_2015 = data[[5]]; data_2015
```

```python
data_2015[5].astype(int)
```

```python
top_square = data_2015.sort_values(5, axis=0, ascending = False).head(16); top_square
```

```python
top_square_province = top_square.tail(10)
top_square_province.columns = ['Dân số năm 2015']
top_square_province
```

```python
top_square_province.plot.bar()
```
