```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pi = np.pi
x = np.linspace(-2*pi, 2*pi, num=50)
plt.plot(x, np.sin(x))
plt.legend(['sin(x)'])
```

```python
plt.plot(x, np.cos(x))
plt.legend(['cos(x)'])
```

```python
plt.plot(x, np.tan(x))
plt.legend(['tan(x)'])
```
