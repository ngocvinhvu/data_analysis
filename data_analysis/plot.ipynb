```python
"""Vẽ đồ thị hàm x = y**2 và x = y trên cùng 1 trục toạ độ."""
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.axis([0, 20, -20, 20])
x = np.linspace(0, 20, num=50)
plt.plot(x, np.sqrt(x))
plt.plot(x, -np.sqrt(x))
plt.plot(x, x)
plt.legend(['x = y**2', 'x = y**2', 'x = y'])
```
